### Проверка расчетных параметров

Проверим правильность расчета эффективности обогащения на обучающей выборке:

```python
def calculate_rougher_output_recovery(row):
    C = row['rougher.output.concentrate_au']
    F = row['rougher.input.feed_au']
    T = row['rougher.output.tail_au']
    result = 100 * C * (F - T) / (F * (C - T))
    return result
```

```python
train_rougher_output_recovery = df_train.apply(calculate_rougher_output_recovery, axis=1)
mean_absolute_error(df_train['rougher.output.recovery'], train_rougher_output_recovery)
```

Средняя абсолютная ошибка крайне мала, а значит, отклонение несущественное: можно принять, что эффективность рассчитана верно.

### Обработка пропусков

Поскольку все наблюдения отсортированы по времени, разумно воспользоваться *ffill()* или *interpolate()* для заполнения недостающих данных.

```python
def fill_missing_data (df):
    return df.interpolate()
```

Для наших моделей обучения столбец `date` не содержит ничего полезного, однако может пригодиться для углубленого анализа. В качестве компромиссного варианта превратим его в индекс для наших таблиц:

```python
def date_to_index (df):
    result = df.copy()
    try:
        result['date'] = pd.to_datetime(df['date'])
        result = df.set_index('date')
    except:
        print ('Конверсия дат не удалась! Проверьте корректность столбца "date".')
    return result
```

Проведем нашу предобработку:

```python
def preprocess(df):
    return (df.
            pipe(fill_missing_data).
            pipe(date_to_index)
           )
```

```python
df_train = preprocess(df_train)
df_test = preprocess(df_test)
df_full = preprocess(df_full)
```

## Модели машинного обучения

### Реализация метрики

Первым делом реализуем метрику sMAPE согласно проектной документации:

$$\mathit{sMAPE} = \frac{1}{N} \sum_{i=1}^{N} \frac{\left| y_i - \hat{y}_i \right|}{\left( \left| y_i \right| + \left|\hat{y}_i \right| \right) / 2} \times 100\%$$

$$\mathit{Итоговое\ sMAPE} = 25\% \times \mathit{sMAPE}\left( \mathit{rougher} \right) + 75\% \times \mathit{sMAPE} \left( \mathit{final} \right)$$

```python
def smape(target, predictions):
    result = np.mean (abs(target - predictions) / ((abs(target)+abs(predictions))/2)) * 100
    return result
```

```python
def adjusted_smape(target,predictions):
    smape_rougher = smape(target['rougher.output.recovery'], predictions[:,0])
    smape_final = smape(target['final.output.recovery'], predictions[:,1])
    return 0.25 * smape_rougher + 0.75 * smape_final
```

Большинство регрессоров sklearn могут работать с вектором целевых признаков, поэтому в качестве целевой метрики мы будем использовать именно итоговое sMAPE (а для регрессоров, не умеющих в мультитаргет, мы воспользуемся оберткой *sklearn.multioutput.MultiOutputRegressor()*).

```python
smape_score = make_scorer(adjusted_smape, greater_is_better=False)
```

### Разбиение выборок

Мы не будем выделять отдельной выборки для валидации и воспользуемся *GridSearchCV* для автоматического тюнинга и кросс-валидации моделей.

```python
# Обновим список общих столбцов тестовой и обучающей выборки:
column_list = list(set(df_train.columns).intersection(set(df_test.columns)))

features_train = df_train[column_list]
target_train = df_train[['rougher.output.recovery', 'final.output.recovery']]
```

Подготовим также признаки для тестовой выборки:

```python
df_test_filtered = df_test[column_list].join(df_full[['rougher.output.recovery','final.output.recovery']], how='inner')
features_test = df_test_filtered.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_test = df_test_filtered[['rougher.output.recovery','final.output.recovery']]
```

```python
# Функция оптимизации модели. Принимает на вход модель, диапазон гиперпараметров для GridSearchCV и обучающую выборку.
# Возвращает модель с оптимальными гиперпараметрами.
def optimize(model, params, X, y):
    name = f'{type(model).__name__}'
    if hasattr(model, 'estimator'):
        name = f'{type(model.estimator).__name__}'
    print(f'Оптимизация {name}...')
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    gcv = GridSearchCV (estimator=model, param_grid=params, cv=5, scoring=smape_score, n_jobs=-1)
    gcv.fit (X_scaled, y)
    print(f'Лучшие параметры: {gcv.best_params_}')
    print(f'sMAPE: {-gcv.best_score_:.2f}')
    return gcv.best_estimator_
```

### Оптимизация гиперпараметров

```python
linear_params = {
                    'fit_intercept':[True, False],
                    'copy_X':[True, False]
                }
best_linear = optimize(LinearRegression(), linear_params, features_train, target_train)
```

```python
tree_params = {
                    'min_samples_leaf': range(1, 7),
                    'max_depth': range(2,8)
              }
best_tree = optimize(DecisionTreeRegressor(random_state=random_state), tree_params, features_train, target_train)
```

```python
forest_params = {                  
                    'max_depth' : range (1,8),
                    'n_estimators' : [5, 10, 20, 50],
                    'min_samples_leaf' : range (1,6),
                    'max_features' : range(1,7)
                 }
best_forest = optimize(RandomForestRegressor(random_state=random_state), forest_params, features_train, target_train)
```

```python
xgb_params = {  
                'estimator__max_depth' : range(1,5),
                'estimator__n_estimators' : [10, 20],
                'estimator__min_child_weight' : range(1,5),
                'estimator__objective': ['reg:squarederror']
             }
best_xgb = optimize(MultiOutputRegressor(XGBRegressor(random_state=random_state)),
                    xgb_params, features_train, target_train)
```

```python
svr_params = {
                'estimator__epsilon':[0.1, 0.2]
             }

best_svr = optimize(MultiOutputRegressor(SVR()), svr_params, features_train, target_train)
```

### Проверка на тестовой выборке

```python
def test(model, X, y):
    name = f'{type(model).__name__}'
    if hasattr(model, 'estimator'):
        name = f'{type(model.estimator).__name__}'
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    predictions = model.predict(X_scaled)
    print (f'Модель: {name}, итоговое sMAPE {adjusted_smape(y, predictions):.1f}')
```

```python
for model in [best_linear, best_tree, best_forest, best_xgb, best_svr]:
    test(model, features_test, target_test)
```

Осталось проверить, насколько результат лучше фиктивной модели. Для этого воспользуемся *DummyRegressor()* из sklearn.dummy.

```python
for strategy in ['mean', 'median']:
    dummy = DummyRegressor(strategy=strategy)
    scaler = StandardScaler()
    dummy_train_scaled = scaler.fit_transform(features_train)
    features_test_scaled = scaler.transform(features_test)
    dummy.fit(dummy_train_scaled, target_train)
    dummy_predictions = dummy.predict(features_test_scaled)
       
    dummy_smape = adjusted_smape (target_test, dummy_predictions)
    
    print(f'Стратегия: {strategy}, итоговое SMAPE: {dummy_smape:.1f}.')
```

Увы, хотя лучшие из наших моделей (Random Forest и SVR, XGBoost показывает разные результаты в зависимости от версии) и обходят фиктивную на тестовой выборке, столь несущественная разница метрики едва ли оправдывает затраты машинного времени.

### Стейкинг

Сильную дисперсию модели можно попробовать компенсировать ансамблированием. Неплохой вариант - объединить разнородные модели (например, линейную регрессию и SVR) и обучить простую метамодель на их предсказаниях:

```python
def refit_and_test(model):
    scaler = StandardScaler()
    scaler.fit(features_train)
    features_train_scaled = scaler.transform(features_train)
    features_test_scaled = scaler.transform(features_test)
    model.fit(features_train_scaled, target_train)
    predictions = model.predict(features_test_scaled)
    print (f'Модель: {type(model.estimator).__name__}, итоговое sMAPE {adjusted_smape(target_test, predictions):.1f}')
```

```python
stack = MultiOutputRegressor(StackingRegressor(estimators=[('SVR', best_svr.estimator), ('linear', best_linear)],
                                               final_estimator=LinearRegression(), cv=5, n_jobs=-1))
refit_and_test(stack)
```

Этот простой способ позволил нам обойти константную модель по метрике уже более уверенно - аж почти на целый процент.

### Голосование

```python
voters = [('SVR', best_svr.estimator), ('linear', best_linear), ('forest', best_forest)]

vote = MultiOutputRegressor(VotingRegressor(estimators=voters, n_jobs=-1))
refit_and_test(vote)
```

В ходе предобработки мы постарались исключить все признаки с высоким VIF, какие могли без ущерба для модели. В основном они относятся к параметрам состояния этапа.

### Рекомендации по заполнению данных

В целом данные удовлетворительны для собранных в автоматическом режиме. Пропущенные значения нетрудно заполнить в силу периодического характера наблюдений.

Некоторого прояснения требуют нулевые концентрации компонентов: они могут быть обусловлены особенностями процесса, отсутствием данных или комбинацией этих факторов.

К сожалению, собранные признаки не позволяют прогнозировать результат существенно лучше, чем усреднение. Большинство признаков слабо связаны с целевыми, но очень сильно - между собой. Скорее всего, сама методика сбора данных нуждается в пересмотре.
