# Исследование закономерностей успеха компьютерных игр

Из открытых источников взяты исторические данные о продажах видеоигр, оценки пользователей и экспертов, жанры и игровые платформы. Нужно выявить закономерности, определяющие успешные продажи для глобального онлайн-магазина. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Предоставленные данные:**
- выборка данных по продажам видеоигр до 2016 года (в том числе жанр, год выпуска, объемы продаж по регионам, рейтинги).

**Этапы работы:**
1. Обзор предоставленных данных.
2. Подготовка данных: исправление ошибок, заполнение пропусков и приведение типов.
3. Исследовательский анализ данных:
    - количество релизов в разные годы;
    - платформы с наибольшими суммарными продажами и распределение по годам;
    - за какой характерный срок появляются новые и исчезают старые платформы?
    - выделение актуального периода для построения прогноза на 2017 год;
    - определение потенциально прибыльных платформ;
    - диаграммы размаха по глобальным продажам игр в разбивке по платформам;
    - влияние на продажи оценок пользователей и критиков;
    - распределение игр по жанрам, определение самых прибыльных жанров.
4. Портрет пользователя по регионам (NA, EU, JP).
    - топ-5 популярных платформ;
    - топ-5 популярных жанров;
    - влияение рейтинга ESRB на продажи в регионе.
5. Проверка гипотез:
    - гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    - гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.
6. Общий вывод.

Смотрим далее, при беглом осмотре оказывается, что у тайтлов есть куча примечаний в скобках. В основном нас интересуют те, что касаются продаж (sales):

```python
def parse_name (row):
    name = row['name']
    try:
        # Будем искать слово sales внутри скобкок:
        match = re.search(r'\s?\(.*?sales.*?\)', name, flags=re.IGNORECASE)
        if match:
            return True
        else:
            return False
    
    except:
        print ('На входе не строка!')
        return False

df[df.apply(parse_name, axis=1)]['name'].sort_values().tolist()
```

Выглядит крайне плачевно, нетрудно убедиться, что качество приписок практически мусорное, так, строка, отмеченная 'JP sales' может иметь данные по всем продажам, кроме собственно JP. Отдельно радуют old sales, которые больше, чем обычные, и correct sales (остальные, видимо, все incorrect, расходимся). Наконец, имеются weekly sales, мало уступающие обычным - видимо, имеются в виду продажи первой недели, которые действительно зачастую составляют немалую долю от общих.

Создадим отдельный столбец для отфильтрованных от этих комментариев названий:

```python
def strip_name (row):
    name = row['name']
    try:
        parts = re.split(r'\s?\(.*?sales.*?\)', name, flags=re.IGNORECASE)
        if parts[0]:
            return parts[0].lower().replace(':','')
        else:
            return name.lower().replace(':','')
        
    except:
        print ('На входе не строка!')
        return name

df['short_name'] = df.apply(strip_name, axis=1)
```

Что касается прочих совпадений, сгруппируем таблицу по параметрам:
- название без комментариев;

и воспользуемся максимальным значением в группе. В некоторых случаях, возможно, было бы разумнее суммировать, но качество данных не позволяет нам судить об этом однозначно.

```python
# Возьмем максимум по продажам:
title_grouping = df.groupby(['short_name', 'platform', 'year_of_release'])
column_list1 = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
df[column_list1] = title_grouping[column_list1].transform ('max')

# И первое непустое значение для категорий:
column_list2 = ['genre', 'critic_score', 'user_score', 'rating']
df[column_list2] = title_grouping[column_list2].transform ('first')

# Полный тайтл нам больше не понадобится:
df['name'] = df['short_name']
df = df.drop('short_name', axis=1)
df = df.drop_duplicates().reset_index(drop=True)
```

Распределение оценок критиков выглядит разумно, без аномальных пиков и нулей, однако около половины значений пропущено. Пользовательских оценок также пропущено в районе 40%, плюс к этому есть и значения 'tbd', которые с точки зрения нашего исследования ничем не отличаются от пропусков.

В целях унификации рейтингов:
- заменим пропущенные значения всех видов на -1;
- умножим пользовательский рейтинг на 10 и приведем оба рейтинга к целому типу.

```python
df['critic_score'] = df['critic_score'].fillna(-1).astype('int8')
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'] = df['user_score'].astype('float')
df['user_score'] = df['user_score']*10
df.query('user_score == 0')
```

После промежуточного приведения типа одна из строк оказалась с нулевым пользовательским рейтингом. Что ж, мы не знаем критериев составления рейтинга, по крайней мере убедимся, что распределение оценок разумное:

```python
df['user_score'].hist().set(title='Распределение пользовательских оценок', xlabel='Оценки пользователей', ylabel='Количество, шт');
```

Теперь закончим заполнение пропусков и приведем столбец к целому:

```python
df['user_score'] = df['user_score'].fillna(-1).astype('int8')
```

И наконец, рассмотрим рейтинг ESRB:

```python
df['rating'].value_counts()
```

Согласно esrb.org, рейтинги EC, E, E10+, T, M и AO являются действующими, RP означает *rating pending* (рейтинг еще не проставлен). Рейтинг K-A - старое обозначение рейтинга Е, так что проведем соответсвующую замену:

```python
df['rating'] = df['rating'].replace('K-A', 'E')
```

Наличие рейтинга RP должно нас насторожить, поскольку согласно правилам ESRB, он может использоваться только до официального релиза:

```python
df.query('rating == "RP"')
```

Продажи есть, а рейтингов все нет. Сайт ESRB, впрочем, подсказывает, что реальный рейтинг для Supreme Ruler: Cold War - E10+, а для WRC: FIA World Rally Championship - E.

Что касается Clockwork Empires, игра по сути умерла, не выйдя из раннего доступа, так что отсутствие конечного рейтинга в данном случае вполне правомерно.

```python
df.loc[df['name'] == "Supreme Ruler: Cold War", 'rating'] = 'E10+'
df.loc[df['name'] == "WRC: FIA World Rally Championship", 'rating'] = 'E'
```

Другие пропущенные рейтинги можно было бы стащить откуда-нибудь по API, но все же выгрузка данных - не наша работа, так что воспользуемся тем, что дают.

Все оставшиеся неизвестные значения заменим на `N/A` и приведем столбец к категориальному типу:

```python
df['rating'] = df['rating'].replace('RP', 'N/A')
df['rating'] = df['rating'].fillna('N/A').astype('category')
```

### Подсчет суммарных продаж 

Суммирование трех столбцов - более чем тривиальная задача. Тем не менее, воспользуемся моментом и приведем типы к float32 для экономии памяти:

```python
df['na_sales'] = df['na_sales'].astype('float32')
df['eu_sales'] = df['eu_sales'].astype('float32')
df['jp_sales'] = df['jp_sales'].astype('float32')
df['other_sales'] = df['other_sales'].astype('float32')
df['world_sales'] = df.loc[:, ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
df.sample(n=5, random_state=177013)
```

## Статистический анализ данных

### Рейтинги платформ

Нам предстоит проверить гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, или, говоря строго:
- *нулевая гипотеза* (H<sub>0</sub>): средние пользовательские рейтинги платформ Xbox One и PC равны;
- *альтернативная гипотеза* (H<sub>1</sub>): средние пользовательские рейтинги платформ не равны.

Это типичная гипотеза о равенстве среднего двух генеральных совокупностей по выборкам из них. Выборки независимы, поэтому воспользуемся соответствующим T-тестом для двух независимых выборок (*ttest_ind()*). Уровень значимости критерия Стьюдента примем в 5%

Мы не будем проводить тесты на равенство дисперсий как нецелесообразные (см. например *Zimmerman, D. W. (2004). A note on preliminary tests of equality of variances. British Journal of Mathematical and Statistical Psychology 57(1): 173–181*) и сразу воспользуемся тестом Уэлча (`equal_var=False`).

Отметим, что будем работать с выборкой за **актуальный период**.

```python
df_xone = df_recent.query('user_score != -1 and platform == "XOne"')
df_pc = df_recent.query('user_score != -1 and platform == "PC"')

alpha = 0.05
results1 = st.ttest_ind (df_xone['user_score'], df_pc['user_score'], equal_var=False) 

print ('P-значение:', results1.pvalue)

if (results1.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Похоже, средние рейтинги действительно близки, по крайней мере, с 2014 по 2016 годы.

```python
df_recent.query('user_score != -1').pivot_table(index='platform', values='user_score', aggfunc=['mean','median'])
```

Различия действительно невелики, и с существенной вероятностью обусловлены случайностью (p-значение около 0,12).

### Рейтинги жанров

Средние пользовательские рейтинги жанров Action и Sports разные.

Следующая гипотеза о том, что средние пользовательские рейтинги жанров Action и Sports разные, иными словами:
- нулевая гипотеза (H0): средние пользовательские рейтинги жанров Action и Sports равны;
- альтернативная гипотеза (H1): средние пользовательские рейтинги не равны.

Это также гипотеза о равенстве среднего двух генеральных совокупностей по выборкам, и выборки также независимы. Уровень значимости примем в 0,1%.

Как и прежде, будем работать с выборкой за **актуальный период**.

```python
df_action = df_recent.query('user_score != -1 and genre == "Action"')
df_sports = df_recent.query('user_score != -1 and genre == "Sports"')

alpha = 0.001
results2 = st.ttest_ind (df_action['user_score'], df_sports['user_score'], equal_var=False) 
print ('P-значение:', results2.pvalue)

if (results2.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

P-значение в данном случае очень мало. Средний рейтинг по жанру Action в нашей выборке около 70, а Sports в районе 55 - разница большая, и по-видимому, это не случайно:

```python
df_recent.query('user_score != -1').pivot_table(index='genre', values='user_score', aggfunc=['mean','median'])
```

### Вывод

Мы проверили две нулевых гипотезы:

- средние пользовательские рейтинги платформ Xbox One и PC равны;
- средняя пользовательские рейтинги  жанров Action и Sports равны.

В качестве альтернативных гипотез были выбраны утверждения, что соответствующие средние пользовательские рейтинги не равны. Проверка с использованием Т-теста для двух независимых выборок показала:

- первая гипотеза не может быть отвергнута даже с относительно большим уровнем значимости 5%. Обнаруженное расхождение в нашей выборке также невелико.
- вторая гипотеза легко опровергается с уровнем значимости 0,1%. Разница по нашей выборке существенна и, по-видимому, это не простое совпадение - средние пользовательские рейтинги жанров Action и Sports действительно разные.

- Игры для старшей аудитории создают больше продаж в Европе и Америке, в Японии аудитория более подростковая. Игры для всех возрастов стабильно продаются везде. Рейтинг ESRB при планировании в Японии, впрочем, следует применять с оглядкой: большинство продаж приходится на игры без рейтинга.

### Рекомендации по выгрузке данных

С несуммированными строками и разрозненными комментариями прямо в столбце с названиями нужно что-то делать, возможно, вести учет продаж по регионам в разных таблицах, - ведь по отдельности данные строк вполне качественные. Разбирать подобную мешанину вручную слишком трудоемко.

Пропуски в столбцах с годом выпуска и рейтингом можно было бы заполнить, обращаясь к какой-либо общедоступной базе игр по API.
