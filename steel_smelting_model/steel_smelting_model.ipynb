Для удобства дальнейшего анализа сразу приведем временные отметки к правильному типу:

```python
def fix_datetime(df, columns):
    try:
        for column in columns:
            df[column] = pd.to_datetime(df[column])
    except:
        print ('Конверсия дат не удалась! Проверьте корректность указания столбцов.')
    return df
```

```python
df_arc = (
    df_arc
    .pipe(fix_datetime, ['Начало нагрева дугой', 'Конец нагрева дугой'])
)
```

```python
def visualize(column, xlim=(0)):
    xlabel = column.name
    fig, axes = plt.subplots(1, 2, figsize=(16,7))
    
    sns.histplot (column, ax=axes[0], bins='fd', kde=False)
    axes[0].set (title='Гистограмма', xlabel=xlabel, ylabel='Наблюдений, шт.', xlim=xlim);
    sns.violinplot (data=column, ax=axes[1], x=column, orient='h')
    axes[1].set (title='Плотность и размах', xlabel=xlabel, xlim=xlim);
```

В большинстве случаев происходит 4-6 итераций нагрева, но есть и одноэтапные партии, и с большим числом "заходов" - до 16. Продолжительность нагрева распределена мультимодально и составлет от 11 с до 10 мин и более, значимая часть по диаграмме размаха не превышает 6 мин. Средняя продолжительность - около 172 с, медиана 147 с.

**При этом оказывается, что мощность линейно зависит от продолжительности итерации, что ставит вопрос о корректной трактовке признака: мощность или энергия?**

Несколько забегая вперед, отметим, что остальные таблицы имеют иную, разреженную структуру: одно вхождение на партию и большое количество столбцов. Для последующего объединения мы можем привести ее к такому же виду, либо просуммировать затраты энергии по всем итерациям, но последнее возможно лишь после разрешения вопроса выше.

Заголовки столбцов даны кириллицей с пробелами, что мы исправим:

```python
def fix_headers(df, new_headers):
    df.columns = new_headers
    return df
```

```python
df_arc = (
    df_arc
    .pipe(fix_headers, ['key', 'start_time', 'end_time', 'active_power', 'reactive_power', 'heat_seconds'])
)
```

Предварительный вариант разреженной таблицы для объединения будет таким (данные о реактивной мощности и длительности исключим ввиду линейной зависимости, но сохраним временные отметки для дальнейшего анализа процесса в целом):

```python
df_arc_sparse = df_arc.pivot_table(index='key',
                   values=['active_power', 'start_time', 'end_time'],
                   columns=df_arc.groupby('key').transform('cumcount'),
                   aggfunc='first')
df_arc_sparse.columns = [''.join(str(s)).strip('()').replace("'",'').replace(', ','_') for s in df_arc_sparse.columns]
df_arc_sparse = df_arc_sparse.reset_index()
df_arc_sparse = (df_arc_sparse
                     .pipe(fix_datetime, list(df_arc_sparse.select_dtypes('object').columns))
                )
for column in df_arc_sparse.filter(like='active_power').columns:
    df_arc_sparse[column] = df_arc_sparse[column].fillna(0)
```

## Подготовка выборок

Переведем временные отметки в секунды со старта, за исключением сильно коррелирующих с мощностью.

```python
def time_to_deltas(df):
    exclude = list(df.filter(like='start_time').columns) + list(df.filter(like='end_time').columns)
    times = df.select_dtypes('datetime').drop(exclude, axis=1)
    for column in times.columns:
        times[column] = times[column].fillna(df['start_time_0'])
        # Если присадки загружались до начала нагрева, будем считать, что это происходило одновременно:
        times[column] = times[column].clip(lower=df['start_time_0'])
        df[column + '_seconds'] = (times[column] - df['start_time_0']).dt.seconds.astype('int')
    
    df = df.drop(df.select_dtypes('datetime').columns, axis=1)
    return df
```

```python
df = df.pipe(time_to_deltas)
```

### Длительность процесса и фильтрация выбросов

Основные подозрения на выбросы приходятся на данные газовой продувки. Поскольку они трактуются как суммарные, имеет смысл также проанализировать общее время выплавки партии.

```python
visualize(df['measure_time_seconds'], (0, 10000))
```

Как видно, в большинстве случаев длительность процесса не превышает часа с небольшим (в среднем 32-36 мин), однако встречаются и очень длительные процессы - до 6,5 ч. Очень малая выборка сверхдлительных процессов дает повод сомневаться в качестве их предсказания. В связи и этим мы попробуем очертить выбросы в пространстве "объем газа - длительность процесса" с помощью расстояния Махаланобиса:

```python
outlier_space = ['gas', 'measure_time_seconds']

# Вычислим матрицу ковариации:
covariance = np.cov(df[outlier_space], rowvar=False)
covariance_pm1 = np.linalg.matrix_power(covariance, -1)
# Найдем центр распределения:
center = np.mean(df[outlier_space], axis=0)
```

```python
def calc_distance(row):
    return spatial.distance.mahalanobis(row, center, covariance_pm1)
```

```python
# Вычислим расстояния для каждого наблюдения:
distance = df[outlier_space].apply(calc_distance, axis=1)
```

Исключим не более 1% наблюдений:

```python
limit = distance.quantile(0.99)
```

```python
# Обрисуем границы выбросов
lambda_, v = np.linalg.eig(covariance)
lambda_ = np.sqrt(lambda_)
ellipse = patches.Ellipse(xy=(center[0], center[1]),
                  width=lambda_[0] * limit * 2, height=lambda_[1] * limit * 2,
                  angle=np.rad2deg(np.arccos(v[0, 0])))
ellipse.set_alpha(0.5)
fig = plt.figure(figsize=(10,6))
ax = plt.subplot()
plt.scatter(x=df['gas'], y=df['measure_time_seconds']);
ax.add_artist(ellipse)
ax.set(xlim=(0), ylim=(0), title='Зависимость объема газа от длительности процесса',
       xlabel='Объем газа', ylabel='Длительность процесса, с');
```

```python
df = df[distance < limit]
df[outlier_space].max()
```

Результат более чем удовлетворительный: теперь максимальный объем газа в таблице порядка 35 единиц (что хорошо согласуется с рекомендованным ограничением), а максимальная длительность плавки чуть более 2 ч.

В этом варианте предобработки мы включим и время загрузки присадок, но при необходимости эту возможность можно отключить.

### Baseline

Определим нижний предел допустимого качества с помощью фиктивной модели: 

```python
for strategy in ['mean', 'median']:
    dummy = DummyRegressor(strategy=strategy)
    dummy.fit(features_train[recommended_columns], target_train)
    dummy_predictions = dummy.predict(features_test[recommended_columns])
       
    mae = mean_absolute_error(target_test, dummy_predictions)
    r2 = r2_score(target_test, dummy_predictions)
    
    print(f'Стратегия: {strategy}, MAE: {mae:.1f} R2: {r2:.1f}')
```

```python
# Покскольку многие признаки распределены со скошенностью, имеет смысл опробовать
# QuantileTransformer()/PowerTransformer() помимо обычного масштабирования.
scaler_list = [MinMaxScaler(),
               StandardScaler(),
               PowerTransformer(),
               QuantileTransformer(random_state=RANDOM_STATE),
               QuantileTransformer(random_state=RANDOM_STATE, output_distribution='normal')]
```

```python
# Функция оптимизации модели. Принимает на вход модель, диапазон гиперпараметров для GridSearchCV и обучающую выборку.
# Возвращает конвейер с оптимальными гиперпараметрами и масштабированием.
def optimize(model, params, X, y):
    name = f'{type(model).__name__}'
    print(f'Оптимизация {name}...')
    pipe = Pipeline([('scaler', StandardScaler()), ('regressor', model)]) 
    gcv = BayesSearchCV(pipe, search_spaces=params, cv=4, scoring='neg_mean_absolute_error', random_state=RANDOM_STATE)
    gcv.fit(X, y)
    print(f'Лучшие параметры: {dict(gcv.best_params_)}')
    print(f'MAE: {(-gcv.best_score_):.2f}')
    return gcv.best_estimator_
```

```python
torch.manual_seed(RANDOM_STATE)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("The model will be running on", device, "device")
```

```python
class Linear(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.fc1 = torch.nn.Linear(n_features, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.relu1 = nn.LeakyReLU()
        self.relu2 = nn.LeakyReLU()
        self.norm1 = nn.BatchNorm1d(64)
        self.norm2 = nn.BatchNorm1d(32)

    def forward(self, batch):
        batch = self.fc1(batch)
        batch = self.relu1(batch)
        batch = self.norm1(batch)
        batch = self.fc2(batch)
        batch = self.relu2(batch)
        batch = self.norm2(batch)
        return self.fc3(batch)
```

```python
module = Linear(stage1_train.shape[1])
```

```python
net = NeuralNetRegressor(
    module,
    criterion = nn.MSELoss,
    optimizer = Adam,
    device = device,
    callbacks = [cp1],
    verbose = 0
)
```

#### Использование признаков

Ввиду разнородности моделей, удобнее будет воспользоваться `sklearn.inspection.permutation_importance()` - этот метод отсортирует признаки по падению качества при перемешивании наблюдений.

```python
model_legend = ['Линейная регрессия', 'Случайный лес', 'Экстремальный лес', 'XGboost', 'SVR', 'Нейросеть']
```

```python
def importance_summary(model_list, X_test, y_test, feature_names):
    importances = pd.DataFrame()
    
    for model in model_list:
        importance = permutation_importance(model,
                                            X_test.astype('float32'),
                                            y_test,
                                            random_state=RANDOM_STATE).importances_mean
        importances = importances.append (pd.Series(importance, index=feature_names),
                                          ignore_index=True)
    
    # Нормализуем показатели, сведя их к одному масштабу:
    importances = importances.div(importances.sum (axis=1), axis=0)
    importances.T.plot(kind='bar', figsize=(12,6), width=0.7).set(title="Влияние признаков в моделях",
                                                                  ylabel="Влияние (нормализовано)");
    plt.legend(model_legend);
```

```python
importance_summary(model_list, stage1_test, target_test, list(features_train[recommended_columns].columns))
```

Картина сходная: затраты электроэнергии и стартовая температура являются решающими факторами, длительность процесса также оказывает заметное влияние (у леса и SVR меньше, у остальных побольше). Объем газа влияет слабее всего.

### Модели с дополнительным feature engineering

Генерация производных признаков зачастую может помочь работе моделей. Производственный процесс к тому же зачастую характеризуется явной физической связью между параметрами. Мы попробуем аппроксимировать эту зависимость с помощью символьной регрессии из комплекта библиотеки генетических аглоритмов gplearn и обучить модели с добавлением нового синтетического признака.

```python
function_set = ['mul', 'div', 'sqrt', 'log', 'inv', 'max', 'min', 'sin', 'cos']
st = SymbolicTransformer(n_components=1, feature_names=recommended_columns,
                           max_samples=0.9,
                           verbose=1,
                           function_set=function_set,
                           n_jobs=-1,
                           random_state=RANDOM_STATE)

st.fit(features_train[recommended_columns], target_train);
```

```python
print(st._best_programs[0])
```

```python
# Если у вас не установлен Graphviz, закомментируйте эту ячейку.
dot_data = st._best_programs[0].export_graphviz()
graphviz.Source(dot_data)
```

Как видим, символьная регрессия достаточно быстро сходится к виду 


$$
T_{end}   \propto   Q \cdot  cos\sqrt{T_{start}}
$$

Эта на первый взгляд странная формула вполне осмысленна: косинус квадратного корня - периодическая функция с локальным максимумом около $T_{start}$=1420 и минимумами около 1200 и 1660, что хорошо согласуется с входным диапазоном температур плавки. При этом чем ниже стартовая точка, тем больше тратится энергии ($Q$) на нагрев, но если температура уже высока, тратить энергию на перегрев нет смысла, более того, перегрев может ухудшить качество стали.

Отметим, что мы можем задать более сложные параметры генетического алгоритма (разрешить сложение и вычитание, увеличить популяцию и т.д.) и получить более сложную схему, в которой формула выше будет одной из ветвей (например, $k \cdot cos\sqrt{T_{start}} + n \cdot Q$). Это улучшит метрику, но ввиду не слишком большой выборки сохранение точности на новых данных сомнительно.
